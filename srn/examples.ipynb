# Examples
We investigate some examples from the paper "Stationary Measures of CTMCS with Applivations to SRNS"

```python
from main import SRN, create_network, plot_network, plot_distribution, plot_generating_terms, plot_gamma
import matplotlib.pyplot as plt
from arviz import style
%reload_ext watermark

style.use(["arviz-darkgrid", "arviz-viridish"])
plt.rcParams["figure.figsize"] = [10, 6]
plt.rcParams["figure.dpi"] = 150
plt.rcParams["figure.facecolor"] = "white"
```

### Example 3.4 (A)

```python
g = create_network(
        reactions=[("0S", "2S", 1), ("2S", "0S", 1), ("5S", "1S", 1)]
    )
s = SRN(g)
s.info()
```

### Example 9.1

```python
g = create_network(
        reactions=[("1S", "2S", 10), ("2S", "3S", 10), ("3S", "2S", 1), ("3S", "1S", 2)]
    )
```

```python
plot_network(g)
```

We create an instance of the SRN class based on the network g and derive the constants:

```python
s = SRN(g)
s.info()
```

```python
pi = s.stationary_distribution(n_max=70)
plot_distribution(pi)
```

```python
plot_generating_terms(s, 20)
```

```python
plot_gamma(s, n=70, scale=True)
```
